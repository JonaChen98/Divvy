# <b> Process <b>
<b> Creating new columns from existing ones: Ride duration(Minutes), ride_distance(station to station), Days of the week service was used <br>
    In the end I did not use the first two columns because exploring the data led me down a different route where this information was not relevant <b>

```python
#Changing string date time to timestamps so we can apply basic arithmetic to them
Yearly['started_at'] = Yearly['started_at'].apply(lambda a:datetime.strptime(a, '%Y-%m-%d %H:%M:%S'))
Yearly['ended_at'] = Yearly['ended_at'].apply(lambda a:datetime.strptime(a, '%Y-%m-%d %H:%M:%S'))
```

```python
#getting the difference in end time and start time
Yearly['ride_duration'] = Yearly['ended_at'] - Yearly['started_at'] 
```

```python
#change from days and minutes to total seconds
Yearly['ride_duration'] = Yearly['ride_duration'].apply(lambda a: a.total_seconds())
```

```python
# Divide the seconds by 60 to convert to minutes
Yearly['ride_duration'] = Yearly['ride_duration']/60
```

```python
# Resulting datframe
Yearly.head(3)
```

<b> Doing reasearch on chicago I came to find that Chicago had some of the worst if not the worst traffic in the United States. With the data I have currently I decided to create maps using coordinates and popular start stations to see if they would conincide with heavy traffic areas. <br> <b>

<b>For my traffic data I used data from Chicago Data Portal, they have a map that shows where taffic numbers daily. <br>
https://data.cityofchicago.org/Transportation/Average-Daily-Traffic-Counts-Map/pf56-35rv <b>

```python
# Creating a Map of the top ten most popular stations for bike rides. We start with a base map of Chicago using Folium
Chicago = folium.Map(location = [41.881832, -87.623177], width = '65%', height = '65%' )
Chicago
```

```python
# Using groupby to create a pivot table that returns data based on the start station name for Memebers
df_locations = Member.groupby(['start_station_name']).agg({"start_station_id":"count", "start_lat":"min", "start_lng":"min"})
df_locations;
```

```python
# Sorting the values from highest to lowest based on number of rides at a station, and returning the top ten
df_locations = df_locations.sort_values(by = "start_station_id", ascending = False)
mtop_10 = df_locations.head(10)
mtop_10
```

```python
# Creating the same pivot table for Casual memebers
cdf_locations = Casual.groupby(['start_station_name']).agg({"start_station_id":"count", "start_lat":"min", "start_lng":"min"})
cdf_locations;
```

```python
cdf_locations = cdf_locations.sort_values(by = "start_station_id", ascending = False)
ctop_10 = cdf_locations.head(10)
ctop_10
```

```python
# Creating list of tuples for folium to plot as markers on the map
m_lat = mtop_10['start_lat']
m_lng = mtop_10['start_lng']
mpop_coordinates = list(zip(m_lat, m_lng))

c_lat = ctop_10['start_lat']
c_lng = ctop_10['start_lng']
cpop_coordinates = list(zip(c_lat,c_lng))
```

<b> With data on the top ten stations for both Casual riders and Members, I plotted the coordinates as markers on the map. I want to confirm that riders are using the service to avoid/beat traffic so what I want to see is that the top ten stations for memebers are around heavy traffic areas while for casual riders I want to see popular stations more spread out and random<b>

```python
for i in range(0,len(mpop_coordinates)):
    folium.Marker(
    location= mpop_coordinates[i],
    icon=folium.Icon(color ="green"),
).add_to(Chicago)
    folium.Marker(
    location= cpop_coordinates[i],
    icon=folium.Icon(color ="red"),
).add_to(Chicago)
```

<b> I used green markers for memebers and red markers for Causal riders <b>

```python
Chicago
```

<b> Comparing the map to the traffic data from Chicago Data Portal, I could see that all ten markers for members (GREEN) were located in heavy traffic areas while markers for casual riders were more spread out and random. This leads me to believe that Members are likely using the service to avoid or beat commute traffic. <b> <br>
    
<b> Next I also made a heat map using every single ride on the table and what it showed supported the map above. More Member usage in heavy traffic areas and evenly spread out usage for Casual riders <b> 

```python
# Creating color palette for different number of rides at coordinate
White_viridis = LinearSegmentedColormap.from_list('white_viridis', [
    (0, '#ffffff'),
    (1e-20,'#440053'),
    (0.2,'#404388'),
    (0.4,'#2a788e'),
    (0.6,'#21a784'),
    (0.8,'#78d151'),
    (1,'#fde624'),
], N=256)

# Creating scatter density plot
def using_mpl_scatter(fig, x, y):
    ax = fig.add_subplot(1,1,1, projection = 'scatter_density')
    density = ax.scatter_density(x,y,dpi = 18,cmap = White_viridis)
    fig.colorbar(density, label = 'points_per_pixel')
    
fig = plt.figure(dpi = 100)
using_mpl_scatter(fig, Member['start_lat'], Member['start_lng'])
plt.title("Member Rides Heat Map")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
```
